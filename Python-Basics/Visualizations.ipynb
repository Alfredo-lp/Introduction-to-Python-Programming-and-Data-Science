```python
import math
x = range(6)
y1 = [5*math.cos(z) for z in x]
y2 = [z**2 for z in x]
y3 = [3*z+2 for z in x]
```

```python
plt.plot(x, y1, 'ro:')
plt.plot(x,y2,'bs--')
plt.plot(x,y3,'g^-.')
plt.legend(["Howard", "O'Hare", "Harlem/Lake"], loc = 'best',
           frameon = False)
plt.grid(True)
plt.show()
```

```python
line = plt.plot(x,y1)
plt.setp(line, color='m', linewidth=2)
```

```python
plt.plot(x,y1, color='m', lw=2)
```

```python
t = range(180)
x1 = [math.sin(2*math.pi*i/60.0) for i in t]
x2 = [math.exp(-i/60.0)*j for i,j in zip(t,x1)]

plt.figure(1)
plt.subplot(2,1,1)
plt.plot(t,x1,'b--')
plt.subplot(2,1,2)
plt.plot(t,x2,'r-')

```

```python
plt.subplot(1,2,1)
plt.plot(t,x2,'b')

plt.subplot(1,2,2)
plt.plot(x,y2,'r')
```

```python
import random
rvs = [random.gauss(100,15) for i in range(500)]
# 500 random numbers from a normal distribution w/ mean 100 & s.d. 15

plt.hist(rvs, bins=50, color='purple', normed=True)
plt.xlabel('Candy')
plt.title('Candygram Histogram')
plt.grid(True)
plt.text(65, .03, 'Too purple?')
plt.show()
```
