```python
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read experimental data and subtract 2 so it matches model

filename = "../../Data/cumulative_adoption.csv"
data_file = open(filename, "r")
data_raw = data_file.readlines()
data = []
for datum in data_raw:
    data.append(float(datum) - 2.0)
```

```python
# Generate synthetic datasets and test whether they are consistent with data

new_plots = plt.figure( figsize = (8, 14) )
plot1 = new_plots.add_subplot( 311 )
plot2 = new_plots.add_subplot( 312 )
plot3 = new_plots.add_subplot( 313 )

x = range(len(data))
x_data = x[:len(data)]

# Do polynomial fit: param[0] is coefficient of largest power in polynomial
params = np.polyfit(x_data, data, 2)
print params
slope_d = params[1]
alpha_d = params[0]*1000

# Get ready to plot accumulation totals
plot1.set_title('Accumulated total')
plot1.set_ylabel("Total accumulation")
plot1.set_xlabel("time [days]")

# Generate N_sets synthetic datasets
N_sets = 100
rate_parameter = 0.0833
slope = []
p_slope = 0
alpha = []
p_alpha = 0

for j in range(N_sets):
    count = 0.0
    series = []

    for i in x:
        count += stats.poisson.rvs(rate_parameter)
        series.append(count)

    plot1.plot(x, series)
    params = np.polyfit(x, series, 2)
    
    # Towards calculating a p-value 
    slope.append(params[1])
    if slope[-1] > slope_d: 
        p_slope += 1
    alpha.append(params[0]*1000)
    if alpha[-1] < alpha_d: 
        p_alpha += 1
 
plot1.plot(x_data, data, linewidth=3)

# Plot distribution of chosen values
plot2.set_xlabel("2nd degree coefficient [x1000]")
plot2.set_ylabel("Probability density" )
plot2.hist(alpha, color = "blue", label = "Poisson", normed = True)
plot2.legend()
plot2.plot([alpha_d, alpha_d], [0, 2.5], color = "black", linestyle = '-', lw = 2)
text2 = "p < " +  "{:6.4f}".format(1.0-p_alpha/float(N_sets))
plot2.text(min(alpha), 2.7, text2, fontsize = 15)

plot3.set_xlabel("1st degree coefficient")
plot3.set_ylabel("Probability density")
plot3.hist(slope, color = "blue", normed = True)
plot3.plot([slope_d, slope_d], [0., 6.5], color = "black", linestyle = '-', lw = 2)
text3 = "p < " +  "{:6.4f}".format(1.0-p_slope/float(N_sets))
plot3.text(min(slope) + 0.005, 9.0, text3, fontsize = 15)

new_plots.savefig("../../Results/out_2nd.eps")
new_plots.show()
```

```python
# Plot density plot of coefficients for Poisson process

synthetic = np.vstack((slope, alpha))
kernel = stats.gaussian_kde(synthetic)

# Regular grid to evaluate kernel density estimation
xmin = synthetic[0,:].min()
xmax = synthetic[0,:].max()
ymin = synthetic[1,:].min()
ymax = synthetic[1,:].max()

X,Y = np.mgrid[xmin:xmax:128j, ymin:ymax:128j]
grid = np.vstack([X.ravel(), Y.ravel()])

Z = np.reshape(kernel(grid), X.shape)

plt.xlabel("1st degree coefficient")
plt.ylabel("2nd degree coefficient [x1000]" )
H, xe, ye, im = plt.hist2d(synthetic[0,:], synthetic[1,:], bins=(50,50), cmap=plt.cm.YlOrRd, normed = True)
cbar = plt.colorbar(im)
con = plt.contour(X, Y, Z, 7, cmap=plt.cm.binary)
cbar.add_lines(con)
plt.scatter(slope_d, alpha_d, c = "black")

plt.savefig("../../Results/out_color.eps")
plt.show()
```
